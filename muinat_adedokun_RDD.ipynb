```
clinicaltrial_2021_rdd1 = sc.textFile('/FileStore/tables/clinicaltrial_2021.csv')
clinicaltrial_2021_rdd1.take(5) 
```

```
header = clinicaltrial_2021_rdd1.first()
```

```
clinicaltrial_2021_rdd1 =  clinicaltrial_2021_rdd1.filter(lambda row: row != header)
```

```
# The number of studies in the dataset. 
print(clinicaltrial_2021_rdd1.distinct().count())
```

```
# Filtering out the header from the RDD
clinical_trial_rdd_no_header = clinicaltrial_2021_rdd1.filter(lambda x: x != header)
```

```
# display the RDD without header
clinical_trial_rdd_no_header.collect()
```

```
clinicaltrial_2021_rdd1.collect()
```

```
# spliting the rdd1 with delimitter "|"
clinicaltrial_2021_rdd2 = clinicaltrial_2021_rdd1.map(lambda line: line.split("|"))
```

```
# extracting the colunm type 
clinicaltrial_2021_rdd3 = clinicaltrial_2021_rdd2.map(lambda pair: (pair[5],1))
```

```
# display the rdd 3 
clinicaltrial_2021_rdd3.collect()
```

```
# reduce transformation
clinicaltrial_2021_rdd4 = clinicaltrial_2021_rdd3.reduceByKey(lambda a,b: a + b)
```

```
# sort the rdd4 by number of trials in descending order
clinicaltrial_2021_rdd5 = clinicaltrial_2021_rdd4.sortBy(lambda pair: -pair[1])
```

```
# You should list all the types (as contained in the Type column) of studies in the 
#dataset along with the frequencies of each type. 
clinicaltrial_2021_rdd5.take(5)
```

```
# Extract Conditions column and split values by comma
rdd6 = clinicaltrial_2021_rdd2.map(lambda x:x[7].split(","))
```

```
rdd6.collect()
```

```
# Flatten the list of condition
conditions_flat_rdd = rdd6.flatMap(lambda x: x)
```

```
conditions_flat_rdd.collect()
```

```
# map each condition to a tuple with a count of 1
condition_counts = conditions_flat_rdd.map(lambda x: (x,1))
```

```
# Reduce the tuples by key to count the frequency of each condition
condition_freq_rdd = condition_counts.filter(lambda x: x[0] != '').reduceByKey(lambda a, b: a + b)
```

```
# Sort the resulting RDD by frequency in descending order
condition_sorted_rdd = condition_freq_rdd.sortBy(lambda x: x[1], ascending = False)
```

```
# take the top 5
top_conditions = condition_sorted_rdd.take(5)
```

```
for condition, count in top_conditions:
     print(condition, count)
```

```
import time
# Filter the RDD to include only trials completed in 2021 and with 'Completed' status

clinicaltrial_rdd8 = clinicaltrial_2021_rdd2.filter(lambda row: row[4][-4:] == '2021' and row[2] == 'Completed')

month_count_rdd = clinicaltrial_rdd8.map(lambda row: (row[4][0:3], 1)).countByKey()

# Sort the dictionary by month
month_count_rdd = dict(sorted(month_count_rdd.items(), key=lambda x: time.strptime(x[0], "%b")))

print(month_count_rdd)


```

```
# analyze the status of each study
# Map each row to a tuple of (status, 1) and reduce by key to get the count for each status
status_counts_rdd = clinicaltrial_2021_rdd2.filter(lambda x: x[2] != "").map(lambda x: (x[2], 1)).reduceByKey(lambda x, y: x + y)
```

```
# Print the status counts
for status, count in status_counts_rdd.collect():
    print(f"{status}: {count}")
```
