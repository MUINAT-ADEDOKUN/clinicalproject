```
clinicaltrial_df = clinicaltrial_df.toDF("Id", "Sponsor", "Status", "Start", "Completion", "Type", "Submission", "Conditions", "Interventions")
clinicaltrial_df.show(4,truncate =False )
```

```
## question 1 The number of studies in the dataset. 
from pyspark.sql.functions import col

num_studies = clinicaltrial_df.filter(col("Id") != "Id").select(col("Id")).distinct().count()
print("Number of distinct studies:", num_studies)
```

```
# Filter out the rows where Type == "Type"
clinicaltrial_df_filtered = clinicaltrial_df.filter(col("Type") != "Type")
```

```
# Group by Type and count the number of rows in each group
type_counts = clinicaltrial_df_filtered.groupBy("Type").count()
```

```
# Order by count in descending order
type_counts = type_counts.orderBy(col("count").desc())
```

```
type_counts.show()
```

```
#top 5 conditions with their frequencies
from pyspark.sql.functions import col, explode, split

top_conditions = (clinicaltrial_df
                  .select(explode(split(col("Conditions"), ",")))
                  .groupBy(col("col").alias("Condition"))
                  .count().filter(col("Condition") != "")
                  .orderBy(col("count").desc())
                  .limit(5))

top_conditions.show()
```

```
from pyspark.sql.functions import col

pharma_df = spark.read.csv("/FileStore/tables/pharma.csv",header = True)



```

```
# select only the distinct parent company
from pyspark.sql.functions import col
parent_companies = pharma_df.select("Parent_Company").distinct()
```

```
# join the clinicaltrial with the parent company 
filtered_clinical_df = clinicaltrial_df.join(parent_companies,clinicaltrial_df["Sponsor"] == parent_companies["parent_company"],'left_anti')
```

```
#counting the trial spons
from pyspark.sql.functions import desc
sponsor_count = filtered_clinical_df.groupBy("Sponsor").count().orderBy(desc("count"))
```

```
# changing the colunm name
sponsor_count = sponsor_count.withColumnRenamed('count', 'number of sponsored Trial')
```

```
# show top 10 sponsor an their sponsored trial
sponsor_count.show(10)
```

```
## Plot number of completed studies each month in a given year – for the submission 
## dataset, the year is 2021.
from pyspark.sql.functions import split, col,substring

# Filter the DataFrame to include only trials completed in 2021 and with 'Completed' status
completed_df = clinicaltrial_df.filter(col("STATUS") == "Completed").filter(col("COMPLETION").endswith("2021"))

# Extract the completion column
new_completed_df = completed_df.select("COMPLETION").filter(col("COMPLETION").isNotNull())


# Extract month (first three letters) from the completion column
completed_trials_month_df = new_completed_df.withColumn("completion_month", substring(col("Completion"), 1, 3))


# Group by month and count the number of trials
month_counts_df = completed_trials_month_df.groupBy("completion_month").count()

# Sort the DataFrame by month
month_count_df = month_counts_df.orderBy(col("completion_month").asc())

month_count_df.display()
```

```
from pyspark.sql.functions import substring, col, udf
from pyspark.sql.types import StringType

# Define a mapping of month names to their corresponding numbers
month_map = {
    'Jan': '01',
    'Feb': '02',
    'Mar': '03',
    'Apr': '04',
    'May': '05',
    'Jun': '06',
    'Jul': '07',
    'Aug': '08',
    'Sep': '09',
    'Oct': '10',
    'Nov': '11',
    'Dec': '12'
}

# Define a UDF to convert month names to numbers using the mapping
month_to_number = udf(lambda month_name: month_map[month_name], StringType())

# Extract month (first three letters) from the completion column
completed_trials_month_df = (
    new_completed_df
    .withColumn("completion_month", substring(col("Completion"), 1, 3))
)

# Add a new column with the month number using the UDF
month_counts_df = (
    completed_trials_month_df
    .withColumn('month_num', month_to_number(col('completion_month')))
    .groupBy('completion_month', 'month_num')
    .count()
    .orderBy('month_num')
    .drop('month_num')
)

# Show the resulting DataFrame
month_counts_df.show()
```

```
## top 10 intervention 
from pyspark.sql.functions import col

# Filter out rows  "Interventions"
condition_filtered_df = clinicaltrial_df.filter(col("interventions") != "interventions") 

# Group the DataFrame by InterventionType and count the number of trials for each type
intervention_type_counts = (
    condition_filtered_df
    .groupBy("Interventions")
    .count().filter(col("interventions") != "")
    .orderBy(col("count").desc())
    .limit(10)
)

# Print the top intervention types for the specified disease
print(f"The top intervention types for are:")
intervention_type_counts.show()




```
